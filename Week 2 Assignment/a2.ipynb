## Question 2

```python
data_Q2 = ren.dropna(subset=['EDUCATION'], inplace = False)
data_Q2.dropna(subset=['WORKTIME'], inplace = True)
X_Q2 = data_Q2[['CREDIT', 'HOUSE', 'CAR', 'EDUCATION', 'WORKTIME']]
y_Q2 = data_Q2['DEFAULT']
X_Q2 = sa.add_constant(X_Q2)
logit = sa.Logit(y_Q2, X_Q2)
res = logit.fit()
res.summary()
```

# Question 4

```python
p2p_real = p2p[['Collapse', 'RegCapital', 'Capitaldeposit', 'Obtaininvest',\
    'Joinasso', 'Autobid', 'Transright', 'Riskdeposit', 'Thirdguarantee']]
p2p['OnlineTime_YMD_new'] = pd.to_datetime(p2p['OnlineTime_YMD'], format='%Y%m%d')
p2p['Bankrupt_WDZJ_new'] = pd.to_datetime(p2p['Bankrupt_WDZJ'], format='%Y%m%d')
p2p_real['Duration'] = (p2p['Bankrupt_WDZJ_new'] - p2p['OnlineTime_YMD_new'])\
    .dt.days
```

```python
p2p_real.dropna(how='any', inplace=True)
p2p_real = p2p_real[p2p_real['Duration'] != 0]
cox = CoxPHFitter()
cox.fit(p2p_real, 'Duration', event_col='Collapse')
cox.print_summary()
```
